## Requirement 4: Compare different bidding algorithms

The different algorithms that we have seen for bidding are considered and we let them play one
against the others in a **generalized first-price auction**.  In particular, consider:
- A primal-dual algorithm for truthful auctions
- A primal-dual algorithm for non-truthful auctions
- A UCB-like approach

We first run the "solo" scenario where we num_participants and they are devided in three equally sized groups:
1. truthful bidders with pacing strategy
2. non-truthful bidders 
3. UCB bidders

In this case defining the regret is not straightforward since we are considering all agents. So what is done is compute all the regrets (w.r.t. adversarial clairvoyant) of the agents and compute the mean for each group and each iteration. For this reason keep in mind that the code scales super poorly with the number of iterations, num_participants and num_auctions. 

```python
%run req4.py --num_auctions 3000 --budget 300 --n_iters 5 --num_participants 30 --eta 0.01 --seed 1 --scenario solo --valuation 0.7
```

Now running the adversarial scenario:

This scenario instead sees 3 agents (1 truthful pacing, 2 non-truthful, 3 UCB) together with num_competitors of adversarial competitors to compete against. In this case we compute just 3 regrets: one for each agent of each type. 

```python
%run req4.py --num_auctions 5000 --budget 200 --n_iters 30 --num_participants 100 --eta 0.01 --seed 1 --scenario adversarial --valuation 0.8 --my_ctrs 0.8,0.8,0.8
```

Let's see what happens when we take the eta learning rate for the truthful bidders to be the one from theory and so much higher than before (we do it by omitting the eta parameter value).

In this case the learning rate is to fast and the bidder converges quickly at bidding close to its valuation (same for all bidders) and dominates the auctions.
